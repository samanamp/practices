## 2. Lecture note (~10 min): vectors & embeddings

**The itch.** To let a computer judge that *dog* and *puppy* are similar, we first must turn things (words, sentences, images) into something math can compare. One number is too poor to capture meaning.

**The picture.** A **vector** is an ordered list of numbers, $x = (x_1, \dots, x_d)$ — picture it as an **arrow from the origin** (or a **point**) in $d$-dimensional space, where each coordinate $x_i$ is one *feature*. We write $x \in \mathbb{R}^d$ ("a list of $d$ real numbers"); $d$ is the **dimension**.

An **embedding** is a vector chosen so that **geometry encodes meaning**: similar things get vectors pointing in similar directions. That one idea powers search, RAG, and recommendation.

**Tiny example.** 2-D toy embeddings, axis 1 = "royalty", axis 2 = "female": king $=(0.9,0.1)$, queen $=(0.9,0.9)$, man $=(0.1,0.1)$, woman $=(0.1,0.9)$. Meaning becomes *direction*: queen $-$ king $=(0,0.8)\approx$ "add female".

**Knobs & effect.** Bigger $d$ = more nuance (real text embeddings are 384–4096-D). The striking effect: you can do *arithmetic on meaning* — the famous **king − man + woman ≈ queen**.

**In the wild.** word2vec/GloVe, sentence-transformers, CLIP. In RAG, every document is stored as an embedding vector.
