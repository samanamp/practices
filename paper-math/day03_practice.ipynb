## 2. Lecture note (~10 min): the dot product

**The itch.** Two embeddings are just two arrows. We want a *single number* summarizing how much they agree.

**The picture.** Walk down both lists together, multiply matching coordinates, and add up the results. If they tend to be big-and-positive in the same places, the total is large; if one is positive where the other is negative, terms cancel.

**The formula** (built straight on day-1's Σ):
$$a\cdot b = \sum_{i=1}^{d} a_i b_i.$$
The result is a single **scalar**, not a vector.

**Knobs & effect.** The **sign** reads off direction agreement: $a\cdot b > 0$ roughly same way, $=0$ perpendicular (unrelated), $<0$ opposed. The **magnitude**, though, also grows when the vectors are simply longer — a wrinkle we'll fix with the norm (day 4) and angle (day 5). Special case worth remembering: $a\cdot a = \sum_i a_i^2$ (a vector dotted with itself = its squared length — tomorrow's topic).

**In the wild.** Similarity scores, a single neuron's output (weights · inputs), projections, and the $QK^\top$ scores inside attention are *all* dot products.
