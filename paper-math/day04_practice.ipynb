## 2. Lecture note (~10 min): the L2 norm

**The itch.** A dot product is big both when vectors align *and* when they are just long. To compare *directions* we need to measure and divide out length.

**The picture.** In 2-D, a vector $(3,4)$ is the hypotenuse of a right triangle — length $\sqrt{3^2+4^2}=5$. Pythagoras works in any dimension.

**The formula.** The **L2 norm** (length, magnitude) is
$$\lVert x\rVert = \sqrt{\sum_{i=1}^{d} x_i^2} = \sqrt{x\cdot x}.$$
It's literally the square root of yesterday's `self_dot`.

**Knobs & effect.** Scaling a vector scales its length: $\lVert c\,x\rVert = |c|\,\lVert x\rVert$. And $\lVert x\rVert \ge 0$, equal to 0 only for the all-zeros vector (which therefore has *no direction* — it'll be the one input we can't normalize). The **distance** between two points is the length of their difference, $\lVert a-b\rVert$.

**In the wild.** Normalizing embeddings, L2 regularization / weight decay, nearest-neighbor distances, gradient clipping.
