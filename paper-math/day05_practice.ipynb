## 2. Lecture note (~10 min): dot product = lengths × cosine of the angle

**The itch.** The dot product still entangles *how aligned* with *how long*. We want to peel off a number that depends **only on the angle** between two arrows.

**The picture.** Two arrows from the origin form an angle $\theta$. Geometry gives the beautiful identity:
$$a\cdot b = \lVert a\rVert\,\lVert b\rVert\,\cos\theta.$$
The dot product is "(length of $a$) times (length of $b$) times (cosine of the angle)." Rearranged, the angle pops out:
$$\cos\theta = \frac{a\cdot b}{\lVert a\rVert\,\lVert b\rVert}.$$

**Knobs & effect.** $\cos\theta \in [-1, 1]$ regardless of lengths: $\theta=0^\circ\Rightarrow\cos=1$ (same direction), $90^\circ\Rightarrow 0$ (perpendicular), $180^\circ\Rightarrow -1$ (opposite). Because the lengths sit in the denominator, **stretching either vector doesn't change $\cos\theta$** — it's purely about direction.

**In the wild.** This ratio, named **cosine similarity**, is exactly how RAG ranks documents (day 7) and is the core of an attention score (Milestone 2).

```python
def ref_get_dimension(x):
    return len(x)

def ref_norm(x):
    return np.sqrt(np.sum(x**2))

def ref_cosine_of_angle(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def ref_angle_degrees(a, b):
    c = np.clip(ref_cosine_of_angle(a, b), -1.0, 1.0)
    return np.degrees(np.arccos(c))

def ref_reconstruct_dot(a, b):
    return np.linalg.norm(a) * np.linalg.norm(b) * ref_cosine_of_angle(a, b)

check_r1(ref_get_dimension)
check_r2(ref_norm)
check_e1(ref_cosine_of_angle)
check_e2(ref_angle_degrees)
check_e3(ref_reconstruct_dot)
print("\nAll reference solutions pass. 🎉  See you on day 6!")
```
