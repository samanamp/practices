## 2. Lecture note (~10 min): normalization

**The itch.** We keep wanting *direction only*, but raw vectors carry length too. Simplest fix: rescale every vector to length exactly 1.

**The picture.** Take any arrow and shrink/grow it until it just touches the **unit sphere** (the set of length-1 vectors). Direction preserved, length thrown away.

**The formula.** The **unit vector** ("normalized" / "$\hat x$") is
$$\hat x = \frac{x}{\lVert x\rVert}, \qquad \lVert \hat x\rVert = 1.$$

**Knobs & effect.** The payoff: for unit vectors, $\hat a\cdot\hat b = \cos\theta$ directly — normalization turns the plain dot product *into* the angle similarity. Caveat: the zero vector has $\lVert x\rVert=0$ and **cannot be normalized** (division by zero) — it has no direction.

**In the wild.** Many retrieval systems pre-normalize all embeddings so a fast dot product equals cosine similarity; you'll also see it in attention variants and in $\ell_2$-normalized feature heads.

```python
def ref_dot(a, b):
    return float(np.sum(a * b))

def ref_cos(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def ref_normalize(x):
    return x / np.linalg.norm(x)

def ref_normalized_dot(a, b):
    return np.dot(ref_normalize(a), ref_normalize(b))

def ref_normalize_rows(M):
    return M / np.linalg.norm(M, axis=1, keepdims=True)

check_r1(ref_dot)
check_r2(ref_cos)
check_e1(ref_normalize)
check_e2(ref_normalized_dot)
check_e3(ref_normalize_rows)
print("\nAll reference solutions pass. 🎉  See you on day 7!")
```
