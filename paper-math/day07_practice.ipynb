## 2. Lecture note (~10 min): cosine similarity

**The itch.** Search needs *one* score per (query, document) pair that means "how related," and that ignores irrelevant differences in vector length.

**The picture.** The angle between two embeddings: pointing the same way = very similar; perpendicular = unrelated; opposite = anti-related.

**The formula** (just yesterday's pieces, named):
$$\operatorname{cossim}(a,b) = \frac{a\cdot b}{\lVert a\rVert\,\lVert b\rVert} = \hat a\cdot\hat b \in [-1, 1].$$

**Knobs & effect.** It is **symmetric** ($\operatorname{cossim}(a,b)=\operatorname{cossim}(b,a)$) and **scale-invariant** ($\operatorname{cossim}(a,b)=\operatorname{cossim}(ca,b)$ for $c>0$): only direction matters. That is exactly why RAG prefers cosine over raw Euclidean distance — a document shouldn't rank higher just because its embedding happens to be longer.

**In the wild.** Semantic search / RAG retrieval, recommendation, clustering, dedup — and, with normalized $Q,K$, the heart of "cosine attention."
