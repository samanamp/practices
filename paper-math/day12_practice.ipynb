## 2. Lecture note (~10 min): the exponential

**The itch.** To turn arbitrary scores into *weights* we need them all **positive**, and we'd like a bigger score to count for *disproportionately* more. Plain numbers can be negative; we need a transform.

**The picture.** $e^x$ (with $e\approx 2.718$) is a curve that is **always positive**, passes through $e^0=1$, **shrinks toward 0** for negative $x$, and **shoots up** for positive $x$. It's strictly increasing, so it never changes the ranking.

**The key identity.** $e^{a+b}=e^a\,e^b$. So **adding a constant to the input multiplies the output** by a fixed factor — a fact we'll cash in for softmax stability tomorrow.

**Knobs & effect.** *Amplification:* if two inputs differ by $\Delta$, their exponentials differ by the **ratio** $e^{\Delta}$. A gap of 3 becomes a $\approx 20\times$ ratio; a gap of 10, a $\approx 22000\times$ ratio. The flip side is **overflow**: $e^{1000}$ is `inf` in floating point — which is why naive softmax needs a stability trick (day 14).

**In the wild.** softmax, the sigmoid $1/(1+e^{-x})$, Boltzmann/Gibbs distributions in physics, and exponential decay everywhere.
